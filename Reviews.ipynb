```python
import json
import re

def load_reviews(filename):
    with open(filename, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

positive_keywords = {
    "good", "great", "amazing", "excellent",
    "fantastic", "love", "enjoyed", "wonderful", "effective",
    "worth", "solid", "refreshing", "incredible", "perfect",
    "catchy", "upbeat", "polished", "impressed", "favourite",
    "deserved", "revolutionary", "enjoyable", "entertain", "flawless",
    "interesting", "curious", "underrated", "vibrant"

}

negative_keywords = {
    "bad", "terrible", "boring", "awful",
    "hate", "poor", "worst", "disappointing",
    "long", "disaster", "horrible", "lackluster", "suffer",
    "lifeless", "repetitive", "cheap", "bland", "unseasoned",
    "generic", "underwhelming", "uninspired", "uninteresting", "mid", "generic",
    "trainwreck", "disaster", "annoying", "suck", "sucks", "disappointing",
    "cheap", "bland", "lifeless", "empty", "weak", "miss"

}

positive_patterns = [
r"love\w*",
r"amaz\w*",
r"fantastic",
r"great",
r"excellent",
r"fun\w*",
r"worth\w*",
r"perfect\w*"
]

negative_patterns = [
r"bad\w*",
r"\w*horr\w*",
r"awf\w*",
r"annoy\w*",
r"terribl\w*",
r"pain\w*",
r"suffer\w*",
r"borin\w*",
r"repet\w*",
r"underwhelm\w*",
r"uninspir\w*",
r"disappoint\w*",
r"uninterest\w*",
r"lifeless\w*",
r"generic\w*",
r"cheap\w*",
r"bland\w*",
]

stopwords = {
"the", "a", "an", "is", "was", "were",
"and", "to", "of", "in", "it", "that",
"this", "for", "on"
}

def classify_review(text):

    words = text.lower().split()

    pos_score = 0
    neg_score = 0
    temp = ""

    for w in words:
      if w in stopwords:
        words.remove(w)

    for w in words:
      w = re.sub(r'[^\w\s]', '', w)
      if w in positive_keywords:
        if temp == "not" or temp == "isn't" or temp == "doesn't":
          neg_score += 1
        else:
          pos_score += 1
      else:
        for pattern in positive_patterns:
            if re.fullmatch(pattern, w):
                if temp in ("not", "isn't", "doesn't"):
                    neg_score += 1
                else:
                    pos_score += 1
                break

      if w in negative_keywords:
        if temp == "not" or temp == "isn't" or temp == "doesn't":
          pos_score += 1
        else:
          neg_score += 1
      else:
        for pattern in negative_patterns:
            if re.fullmatch(pattern, w):
                if temp in ("not", "isn't", "doesn't"):
                    pos_score += 1
                else:
                    neg_score += 1
                break

      temp = w

    if pos_score > neg_score:
        print("positive")
        return "positive"
    elif neg_score > pos_score:
        print("negative")
        return "negative"
    else:
        print("neutral")
        return "neutral"


def evaluate(data):

    correct = 0
    total = 0

    for item in data:

        text = item["review"]
        true_label = item["label"]

        predicted = classify_review(text)

        if predicted == true_label:
            correct += 1

        total += 1

    accuracy = correct / total

    print("Total reviews:", total)
    print("Correct predictions:", correct)
    print("Accuracy:", round(accuracy, 3))


data = load_reviews("song_reviews.json")

evaluate(data)
```
